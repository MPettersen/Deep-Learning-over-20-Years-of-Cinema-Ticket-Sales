```python
import calendar
import numpy as np
import pandas as pd
from joblib import load

from helper import *
```

```python
showings = load('../data/showings_extra_feature.joblib')
showings.replace(np.inf, 0, inplace=True)
```

# Coverage by day of week

```python
stats = dict()
for i in range(7):
    stats[calendar.day_name[i]] = showings.loc[showings.dayofweek == i, 'coverage']
```

```python
bar_plot(stats, r=0, d=0.1, file_name='day_of_week')
```

```python
stats_outlier = dict()
for i in range(7):
    stats_outlier[calendar.day_name[i]] = showings.loc[(showings.dayofweek == i) & ((showings.hour < 12) | (showings.hour > 22)), 'coverage']
```

```python
bar_plot(stats_outlier, r=0, d=0.1, file_name='day_of_week_outlier')
```
