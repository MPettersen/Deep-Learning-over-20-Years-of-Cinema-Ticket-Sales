```python
showings = load('../data/showings_extra_feature.joblib')
showings.replace(np.inf, 0, inplace=True)
genres = load('../data/genres.joblib')
```

```python
week_of_release = showings.loc[showings.days_since_release < 10, ('dayofweek', 'tmdb_popularity', 'year', 'coverage')]
```

# Coverage by TMDb popularity for the week following the release

```python
popularity = list(set(week_of_release.tmdb_popularity.unique().round()))
popularity.sort()
```

```python
stats = dict()
for i in popularity:
    stats[i] = week_of_release.loc[week_of_release.tmdb_popularity.round() == i, 'coverage']
```

```python
bar_plot(stats, w=20, h=7, r=270, min_cnt=1000)
```
