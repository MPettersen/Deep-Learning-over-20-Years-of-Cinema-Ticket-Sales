```python
import calendar
import numpy as np
import pandas as pd
from joblib import load

from helper import *
```

```python
showings = load('../data/showings_extra_feature.joblib')
showings.replace(np.inf, 0, inplace=True)
genres = load('../data/genres.joblib')
```

# Coverage by four genre combo

```python
%%time
all_genres = showings.tmdb_genre_ids.tolist()
all_genre_tuples = set(tuple(np.sort(i)) for i in all_genres if len(i) > 1)
all_combos = [(i, j, k, l)
              for l in list(genres.keys()) 
              for k in list(genres.keys()) 
              for j in list(genres.keys()) 
              for i in list(genres.keys()) 
              if i < j < k < l]

genre_combo = list()
for i in all_combos:
    for j in all_genre_tuples:
        if i[0] in j and i[1] in j and i[2] in j and i[3] in j:
            genre_combo.append(i)
            break
```

```python
%%time
stats = dict()
for i in genre_combo:
    stats[genres[i[0]] + '-' + genres[i[1]] + '-' + genres[i[2]] + '-' + genres[i[3]]] = showings.loc[
        showings.tmdb_genre_ids.apply(lambda numbers: i[0] in numbers and i[1] in numbers and i[2] in numbers and i[3] in numbers),
        'coverage']
```

```python
bar_plot(stats, r=270, min_cnt=2000)
```
